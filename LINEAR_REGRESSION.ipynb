```python
x=df.iloc[:,:-1]
x.ndim
```

```python
y=df['Salary']
y
```

**SCATTER PLOTING**

```python

plt.xlabel('Year_of_experience',fontsize=10,color='g')
plt.ylabel('Salary',fontsize=10,color='g')
plt.title('Salary Graph',fontsize=20,color='r')
plt.scatter(x,y)
```

```python
#splitting into training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

**MODEL CREATION**

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

**PREDICT THE SALARY**

```python
print(model.predict([[4]]))
```

**REGRESSION LINE**

```python
plt.xlabel('Year_of_experience',fontsize=10,color='g')
plt.ylabel('Salary',fontsize=10,color='g')
plt.scatter(x_train,y_train,color='r')
plt.plot(x_test,y_pred,color='b')
```

**Slope and y=mx+c(Coefficient)**

```python
print("Slope is ",model.coef_)
```

**Intercept**

```python
print("Intercept is",model.intercept_)
```

**MEAN ABSOLUTE ERROR**

```python
from sklearn.metrics import mean_absolute_error
print("Mean Absolute Error",mean_absolute_error(y_test,y_pred))
```

#Mean Absolute percentage error

```python
from sklearn.metrics import mean_absolute_percentage_error
print("mean absolute percentage error",mean_absolute_percentage_error(y_test,y_pred))
```

**MEAN SQUARED ERROR**

```python
from sklearn.metrics import mean_squared_error
print("MEAN SQUARED ERROR:",mean_squared_error(y_test,y_pred))
```

**ROOT_MEAN_SQUARED_ERROR**

```python
MSE=mean_squared_error(y_test,y_pred)
   
rsme = np.sqrt(MSE)  
print("Root Mean Square Error:",rsme)  
```

**R2 SCORE** (Coefficience of determination)



```python
from sklearn.metrics import r2_score
print("r2 score:",r2_score(y_test,y_pred))
```
